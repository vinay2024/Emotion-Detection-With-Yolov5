# Realtime Emotion Detection

```python
cap = cv.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv.imshow('YOLO', np.squeeze(results.render()))
    
    if cv.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv.destroyAllWindows
```
